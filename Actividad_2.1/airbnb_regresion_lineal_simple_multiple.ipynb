# Análisis de Regresión Lineal - Airbnb Cambridge (Estilo simple sin funciones avanzadas)

```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Cargar datos
df = pd.read_csv("Datos_limpios_Cambridge.csv")

# Limpiar columnas relevantes
df["price"] = df["price"].astype(str).str.replace("$", "").str.replace(",", "")
variables = [
    "host_acceptance_rate", "host_response_rate", "review_scores_location",
    "review_scores_cleanliness", "price", "availability_365",
    "number_of_reviews", "reviews_per_month", "review_scores_communication",
    "host_total_listings_count", "accommodates", "bedrooms",
    "review_scores_value"
]

for col in variables:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python

# Crear subconjuntos por tipo de habitación
df_entire = df[df["room_type"] == "Entire home/apt"]
df_private = df[df["room_type"] == "Private room"]
df_shared = df[df["room_type"] == "Shared room"]
df_hotel = df[df["room_type"] == "Hotel room"]
```

## Modelo lineal simple (mayor correlación en Entire home/apt)

```python

#Usamos host_acceptance_rate vs host_response_rate
df_corr = df_entire[["host_acceptance_rate", "host_response_rate"]].dropna()
X = df_corr[["host_response_rate"]]
y = df_corr["host_acceptance_rate"]

model = LinearRegression()
model.fit(X, y)
r2 = model.score(X, y)
print("R²:", r2)
```

```python
sns.scatterplot(x="host_response_rate", y="host_acceptance_rate", data=df_corr)
plt.plot(df_corr["host_response_rate"], model.predict(X), color="red")
plt.title("Modelo simple: host_acceptance_rate vs host_response_rate")
plt.show()
```

## Top 10 relaciones lineales simples por tipo de habitación

```python

from itertools import permutations

vars_modelo = [
    "host_acceptance_rate", "host_total_listings_count", "accommodates",
    "bedrooms", "price", "review_scores_value", "reviews_per_month",
    "number_of_reviews", "availability_365", "review_scores_cleanliness",
    "review_scores_location", "review_scores_communication", "host_response_rate"
]

resultados = []

for dep, indep in permutations(vars_modelo, 2):
    datos = df_entire[[dep, indep]].dropna()
    if datos.shape[0] < 10:
        continue
    X = datos[[indep]]
    y = datos[dep]
    model = LinearRegression()
    model.fit(X, y)
    r2 = model.score(X, y)
    r = np.sqrt(r2)
    resultados.append((dep, indep, r2, r))

top10 = sorted(resultados, key=lambda x: x[2], reverse=True)[:10]

for dep, indep, r2, r in top10:
    print(f"{dep} vs {indep} | R²: {r2:.4f} | r: {r:.4f}")
```
